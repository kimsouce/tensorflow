```python
import tensorflow as tf
import numpy as np
import pandas as pd
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from tensorflow.keras.optimizers import Adam
from sklearn import datasets
# from tensorflow.keras.backend import one_hot #session run이 필요하다. 
from tensorflow.keras.utils import to_categorical #session이 필요 없기 때문에 이걸 쓰면 된다. 굳이 one-hot으로 변경 안해도 된다.
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
xdata = iris['data']
iris['target']
```

```python
ydata = to_categorical(iris['target'],3) #one-hot할 필요 없음. 이게 ydata 이다. 
```

```python
IO = Dense(units=3, input_dim=4, activation='softmax')  #units = 분류의 개수, input_dim = feature의 열의 개수
model = Sequential([IO])
model.compile(loss='categorical_crossentropy',optimizer=Adam(0.1),metrics=['accuracy'])
h = model.fit(xdata,ydata,epochs=1000)
```

```python
IO.get_weights() #W=(4,3), B=3개
```

```python
model.predict_classes(xdata)
```

```python
model.predict( np.array( [[5.1,3.5,1.4,0.2]] ) ) #여기에 예측하고 싶은 값 넣어주는 것이다. #결과값은 hx이다. 
```

```python
h.history['acc'][-1] #맨 마지막 인덱스를 참고하면 된다. 
```
