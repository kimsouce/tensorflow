```python
from keras.models import Sequential
from keras.layers.core import Dense
from tensorflow.keras.optimizers import Adam
from keras.callbacks import EarlyStopping

from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn import model_selection
```

```python
data = pd.read_csv('./dataset/bike_sharing_demand_train.csv')
data.head()
```

```python
xdata = data.iloc[:,:-1]
ydata = data.iloc[:,-1]
```

```python
e = LabelEncoder()
e.fit(ydata)
y = e.transform(ydata)
y_encoded = tf.keras.utils.to_categorical(y)
```

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(xdata,y_encoded,test_size=0.3)
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
model = Sequential() 
model.add(Dense(100, input_dim=10, activation='relu')) 
model.add(Dense(100, activation='softmax')) 
model.add(Dense(822 , activation='softmax')) 
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=100)
```

```python
h = model.fit(x_train, y_train, validation_split=0.2, epochs=300,batch_size= 30, callbacks=[early_stopping_callback])
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(model.predict(x_test),y_test)
```
