```python
import tensorflow as tf
import pandas as pd
import numpy as np

from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('./dataset/sonar.csv', header=None)
data.head()
```

```python
data.info()
```

```python
xdata = data.iloc[:,:-1]
ydata = data.iloc[:,-1]
```

```python
e = LabelEncoder()
e.fit(ydata)
y = e.transform(ydata)
y_encoded = tf.keras.utils.to_categorical(y)
```

```python
IO = Dense(units=2, input_dim=60, activation='sigmoid') #sigmoid를 활성함수로 쓰겠다. 
model = Sequential( [IO] ) #순차모델
model.compile( loss='binary_crossentropy', optimizer=Adam(0.1),metrics=['accuracy'])  #생존 하냐 마냐의 문제이기 때문에 binary 를 쓴다.
h = model.fit( xdata, y_encoded, epochs=200)
```

```python
model.predict_classes(xdata)
```

```python
h.history['accuracy'][-1]
```
