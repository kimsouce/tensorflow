```python
import tensorflow as tf
import pandas as pd
import numpy as np

from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import LabelEncoder
```

```python
iris = pd.read_csv('./dataset/iris.csv', names=['sl','sw','pl','pw','feature_name'])
iris.head()
```

```python
iris.describe() #속성확인1
```

```python
iris.info() #속성확인2
```

```python
iris['feature_name'].unique()  #feature_name에 어떤 종류가 있는지 확인
```

```python
iris = iris.values
```

```python
xdata = iris[:,:-1].astype(float)
ydata = iris[:,-1]
```

```python
ydata
```

```python
e = LabelEncoder()
e.fit(ydata)
y = e.transform(ydata)
y_encoded = tf.keras.utils.to_categorical(y)
```

```python
model = Sequential()
model.add(Dense(16,input_dim=4, activation='softmax'))
model.add(Dense(3, activation='softmax'))
model.compile( loss='categorical_crossentropy', optimizer=Adam(0.1),metrics=['accuracy'])  #생존 하냐 마냐의 문제이기 때문에 binary 를 쓴다.
h = model.fit( xdata, y_encoded, epochs=200, batch_size=1)
```

```python
model.predict_classes(xdata)
```
