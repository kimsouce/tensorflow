```python
from keras.models import Sequential
from keras.layers.core import Dense
from keras.callbacks import EarlyStopping

from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

import numpy
import pandas as pd
import tensorflow as tf
from sklearn import model_selection
```

```python
wine = pd.read_csv('./dataset/wine.csv', header=None)
wine.head()
```

```python
wine.info()
```

```python
wine[12].unique()
```

```python
xdata = wine.iloc[:,:-1]
ydata = wine.iloc[:,[-1]]
```

```python
e = LabelEncoder()
e.fit(ydata)
y = e.transform(ydata)
y_encoded = tf.keras.utils.to_categorical(y)
```

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(xdata,y_encoded,test_size=0.3)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
n_fold = 10
skf = StratifiedKFold(n_splits=n_fold,shuffle=True)
```

```python
accuracy = []
```

```python
for train,test in skf.split(xdata,ydata):
    model = Sequential() 
    model.add(Dense(4, input_dim=12, activation='relu')) 
    model.add(Dense(4, activation='relu')) 
    model.add(Dense(2, activation='sigmoid')) # Compile model 
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) 
```

```python
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=100)
```

```python
model.fit(x_train, y_train, validation_split=0.2, epochs=300,batch_size= 30, callbacks=[early_stopping_callback])
```

```python
print('\n Accuracy: %4f' % (model.evaluate(x_test,y_test)[1]))
```
